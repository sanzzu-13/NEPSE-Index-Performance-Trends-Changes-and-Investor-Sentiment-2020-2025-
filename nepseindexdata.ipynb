# About the Dataset  
## NEPSE Daily Index Data (2020-2025)

This dataset contains daily records of the Nepal Stock Exchange (NEPSE) Index from January 1, 2020, to March 4, 2025. The NEPSE Index reflects the overall performance of Nepalâ€™s stock market by tracking the value of listed companies.

---

## Whatâ€™s in the Dataset?  
- **Date:** The trading day (e.g., 2025-03-04).  
- **Index Value:** The closing value of the NEPSE Index for that day.  
- **Absolute Change:** How much the index changed from the previous day (in points).  
- **Percentage Change:** The daily percentage change (how much the index went up or down).

---

## Data Details  
- Covers over 5 years of daily trading data (excludes weekends and holidays).  
- Contains around 1,200 to 1,500 rows of data.  
- Data sourced from ShareSansar, a popular Nepal financial website.  
- Collected and cleaned using automated Python tools.

# Why Use a Box Plot?
A box plot  is used to summarize the distribution of a dataset. It helps to:

- show the median, minimum, and maximum values

- Display the spread and variability of the data

- Identify outliers (unusual values)

- Compare distributions between variables or over time
##
- The vertical line inside the light blue box shows the median (middle value) of the NEPSE Index, which is around 2000â€“2050.
- The light blue box stretches from about 1900 to 2550, covering the middle 50% of all NEPSE Index values.
- The "whiskers" (lines extending from the box) show the full spread of values (excluding outliers), ranging from about 1200 to 3200.
- There are no dots or points outside the whiskers, meaning there are no extreme values
- The median is slightly to the left, and the right whisker is a bit longer, showing a slight skew toward higher values, but the distribution is mostly balanced.

```python
# Set the plot style
sns.set(style="whitegrid")

# Create a box plot for Absolute Change
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Absolute Change'], color='orange')
plt.title('Box Plot of NEPSE Absolute Change')
plt.xlabel('Absolute Change')
plt.tight_layout()
plt.show()
```

- Shows daily point changes in NEPSE (not percentages).

- Median (middle line) is near zero, meaning most daily changes are small.

- Middle 50% of changes (orange box) range from -15 to +15 points.

- Whiskers stretch from about -70 to +70 points, covering most changes.

- Diamonds outside whiskers are outliersâ€”rare big changes up to Â±150 points.

- The plot is fairly balanced, showing large ups and downs happen equally.



```python
# Create a pair plot
sns.pairplot(df[['Index Value', 'Absolute Change', 'Percentage Change']], diag_kind='kde')
plt.suptitle('Pair Plot for NEPSE Data', y=1.02)
plt.show()
```

- Each scatter plot shows how two variables relate (correlation, trend, clusters, etc.).
- These show the distribution of individual variables.
- Useful to spot groups, trends, or outliers in data.
- Easy way to compare multiple variables at once.
- Helps decide which variables are related or important.

```python
# Compute the correlation matrix
correlation_matrix = df[['Index Value', 'Absolute Change', 'Percentage Change']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for NEPSE Data')
plt.tight_layout()
plt.show()
```

# NEPSE Correlation Matrix

- Correlation values range from -1 to +1:
* +1 = perfect positive relation, -1 = perfect negative, 0 = no relation.

- Each variable is perfectly correlated with itself (value = 1).

- Index Value and Absolute Change: Correlation is almost zero (0.02). This means the current index level doesnâ€™t predict the size of daily point changes.

- Index Value and Percentage Change: Correlation is zero (0.00). So, the index level doesnâ€™t predict daily percent changes either.

- Absolute Change and Percentage Change: Very strong positive correlation (0.97). When point changes are big, percentage changes are big too, and vice versa.
